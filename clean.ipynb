```python
import requests
import os
import cv2
import threading

# Create folder if not exists
folder_path = "filtered_author_images"
os.makedirs(folder_path, exist_ok=True)

# Number of required author-like images
total_authors = 450
downloaded_authors = 0
lock = threading.Lock()

# Load OpenCV face detection model
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")


def is_mature_face(image_path):
    """Check if the image contains a mature-looking face using OpenCV."""
    try:
        img = cv2.imread(image_path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(60, 60))

        # Accept only images with large faces (filtering out children)
        for (x, y, w, h) in faces:
            if h > 100:  # Adjust threshold for better filtering
                return True

    except Exception as e:
        print(f"⚠️ Error processing image: {e}")
    return False


def download_and_filter(index):
    """Download an image, check maturity, and save if valid."""
    global downloaded_authors
    while downloaded_authors < total_authors:
        img_url = "https://thispersondoesnotexist.com/"
        img_path = os.path.join(folder_path, f"temp{index}.jpg")

        try:
            img_data = requests.get(img_url).content
            with open(img_path, "wb") as img_file:
                img_file.write(img_data)

            # Check if image is mature-looking
            if is_mature_face(img_path):
                with lock:
                    if downloaded_authors < total_authors:
                        final_path = os.path.join(folder_path, f"author{downloaded_authors+1}.jpg")
                        os.rename(img_path, final_path)
                        downloaded_authors += 1
                        print(f"✅ Saved author {downloaded_authors}/{total_authors}")

            else:
                os.remove(img_path)  # Delete if not a mature face

        except Exception as e:
            print(f"⚠️ Failed to download image: {e}")


# **Multithreading for faster downloads**
num_threads = 10
threads = []

for i in range(num_threads):
    thread = threading.Thread(target=download_and_filter, args=(i,))
    threads.append(thread)
    thread.start()

for thread in threads:
    thread.join()

print("✅ 450 Mature Author Images Downloaded Successfully!")
```

```python
import requests
import os
import cv2
import threading
import numpy as np

# Create folder if not exists
folder_path = "filtered_author_images"
os.makedirs(folder_path, exist_ok=True)

# Number of required author-like images
total_authors = 450
downloaded_authors = 0
lock = threading.Lock()

# Load OpenCV face detection model
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

def estimate_age_opencv(image_path):
    """Estimate if the face looks 28+ based on facial texture and wrinkles (OpenCV heuristic)."""
    try:
        img = cv2.imread(image_path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Detect faces
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(80, 80))

        if len(faces) == 0:
            return False  # No faces detected

        for (x, y, w, h) in faces:
            face_roi = gray[y:y+h, x:x+w]  # Extract face region
            
            # Compute Laplacian variance (sharpness/wrinkles detection)
            laplacian_var = cv2.Laplacian(face_roi, cv2.CV_64F).var()

            print(f"🔍 Texture Analysis (Laplacian Variance): {laplacian_var} for {image_path}")

            # Heuristic: Faces with higher texture values (~500+) are likely 28+
            if laplacian_var > 500:
                return True

    except Exception as e:
        print(f"⚠️ Error processing image: {e}")
    return False


def download_and_filter(index):
    """Download an image, check if age is 28+ using OpenCV, and save if valid."""
    global downloaded_authors
    while downloaded_authors < total_authors:
        img_url = f"https://thispersondoesnotexist.com/?random={os.urandom(8).hex()}"
        img_path = os.path.join(folder_path, f"temp{index}.jpg")

        try:
            img_data = requests.get(img_url, headers={"Cache-Control": "no-cache"}).content
            with open(img_path, "wb") as img_file:
                img_file.write(img_data)

            # Check if the image contains an adult face (28+ years)
            if estimate_age_opencv(img_path):
                with lock:
                    if downloaded_authors < total_authors:
                        final_path = os.path.join(folder_path, f"author{downloaded_authors+1}.jpg")
                        os.rename(img_path, final_path)
                        downloaded_authors += 1
                        print(f"✅ Saved author {downloaded_authors}/{total_authors}")

            else:
                os.remove(img_path)  # Delete if not a mature face

        except Exception as e:
            print(f"⚠️ Failed to download image: {e}")


# **Multithreading for faster downloads**
num_threads = 10
threads = []

for i in range(num_threads):
    thread = threading.Thread(target=download_and_filter, args=(i,))
    threads.append(thread)
    thread.start()

for thread in threads:
    thread.join()

print("✅ 450 Aged (28+) Author Images Downloaded Successfully!")
```

```python
import requests
import os
import cv2
import threading
import numpy as np

# Create folder if not exists
folder_path = "filtered_author_images"
os.makedirs(folder_path, exist_ok=True)

# Number of required author-like images
total_authors = 450
downloaded_authors = 0
lock = threading.Lock()
image_hashes = set()  # Store unique image hashes

# Load OpenCV face detection model
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")


def get_image_hash(image_path):
    """Generate a perceptual hash for an image."""
    try:
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (32, 32))  # Resize for consistent hashing
        dct = cv2.dct(np.float32(img))  # Apply Discrete Cosine Transform
        return tuple(dct.flatten()[:64])  # Use first 64 values as hash
    except Exception:
        return None


def is_mature_face(image_path):
    """Check if the image contains a mature-looking face using OpenCV."""
    try:
        img = cv2.imread(image_path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(60, 60))

        # Accept only images with large faces (filtering out children)
        for (x, y, w, h) in faces:
            if h > 100:  # Adjust threshold for better filtering
                return True
    except Exception as e:
        print(f"⚠️ Error processing image: {e}")
    return False


def download_and_filter(index):
    """Download an image, check for maturity & uniqueness, and save if valid."""
    global downloaded_authors
    while downloaded_authors < total_authors:
        img_url = "https://thispersondoesnotexist.com/"
        img_path = os.path.join(folder_path, f"temp{index}.jpg")

        try:
            img_data = requests.get(img_url).content
            with open(img_path, "wb") as img_file:
                img_file.write(img_data)

            # Generate hash to check for duplicates
            img_hash = get_image_hash(img_path)
            if img_hash in image_hashes:
                os.remove(img_path)  # Delete duplicate image
                continue

            # Check if image is mature-looking
            if is_mature_face(img_path):
                with lock:
                    if downloaded_authors < total_authors:
                        final_path = os.path.join(folder_path, f"author{downloaded_authors+1}.jpg")
                        os.rename(img_path, final_path)
                        image_hashes.add(img_hash)  # Add to unique image set
                        downloaded_authors += 1
                        print(f"✅ Saved author {downloaded_authors}/{total_authors}")

            else:
                os.remove(img_path)  # Delete if not a mature face

        except Exception as e:
            print(f"⚠️ Failed to download image: {e}")


# **Increased Threads for Maximum Speed**
num_threads = 50  # Increased from 20 to 50
threads = []

for i in range(num_threads):
    thread = threading.Thread(target=download_and_filter, args=(i,))
    threads.append(thread)
    thread.start()

for thread in threads:
    thread.join()

print("✅ 450 Unique & Mature Author Images Downloaded Successfully!")
```

```python
import os
from rembg import remove
from PIL import Image

# Input and Output directories
input_folder = "filtered_author_images"
output_folder = "no_bg_authors"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

def remove_background(image_path, output_path):
    """Removes the background from an image and saves it as PNG."""
    try:
        with open(image_path, "rb") as inp_file:
            img = Image.open(inp_file)
            img = img.convert("RGBA")  # Ensure image has an alpha channel

            # Remove background
            img_no_bg = remove(img)

            # Save the new image
            img_no_bg.save(output_path, "PNG")
            print(f"✅ Processed: {output_path}")

    except Exception as e:
        print(f"⚠️ Error processing {image_path}: {e}")

# Process all images in the folder
for file_name in os.listdir(input_folder):
    if file_name.endswith((".jpg", ".jpeg", ".png")):
        input_path = os.path.join(input_folder, file_name)
        output_path = os.path.join(output_folder, file_name.replace(".jpg", ".png").replace(".jpeg", ".png"))
        remove_background(input_path, output_path)

print("✅ Background removal completed for all images!")
```

```python
import requests
import os
import concurrent.futures
import time

# Define the folder where author images are stored
SAVE_DIR = r"C:\xampp\htdocs\Booksy\author_images"
os.makedirs(SAVE_DIR, exist_ok=True)

# Get all existing images in the folder
existing_images = set(os.listdir(SAVE_DIR))
existing_count = len(existing_images)

# Total required images
TOTAL_IMAGES = 430  
MISSING_IMAGES = TOTAL_IMAGES - existing_count

if MISSING_IMAGES <= 0:
    print(f"✅ You already have {TOTAL_IMAGES} images. No need to download more.")
    exit()

print(f"📌 Found {existing_count} images. Downloading {MISSING_IMAGES} more...\n")

# Function to download an image with retries
def download_image(index, retries=3):
    filename = f"author{index}.jpg"
    file_path = os.path.join(SAVE_DIR, filename)

    # Skip download if file already exists
    if filename in existing_images:
        print(f"✅ {filename} already exists. Skipping...")
        return

    image_url = "https://thispersondoesnotexist.com/image"

    for attempt in range(retries):
        try:
            response = requests.get(image_url, stream=True, timeout=10)
            if response.status_code == 200:
                with open(file_path, 'wb') as file:
                    for chunk in response.iter_content(1024):
                        file.write(chunk)
                print(f"✅ Downloaded: {filename}")
                return  # Exit function after successful download
            else:
                print(f"❌ Failed to download {filename} (Status Code: {response.status_code})")
        except requests.exceptions.RequestException as e:
            print(f"⚠️ Attempt {attempt + 1} failed for {filename}: {e}")

        time.sleep(5)  # Wait before retrying

    print(f"❌ Skipped {filename} after {retries} failed attempts.")

# Use ThreadPoolExecutor for parallel downloads (40 threads)
start_index = existing_count + 1
end_index = start_index + MISSING_IMAGES

with concurrent.futures.ThreadPoolExecutor(max_workers=40) as executor:
    executor.map(download_image, range(start_index, end_index))

print("\n🎉 All missing author images downloaded!")
```
